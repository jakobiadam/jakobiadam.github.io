# Hőterjedés vizsgálata 1 dimenzióban

A mérés célja a hődiffúziós egyenlet kísérleti vizsgálata. A mérési elrendezésben egy réz rúdat használunk, amelynek végeit fűteni, illetve az egyik végét hűteni tudjuk. Cél megvizsgálni milyen gyorsan reagál a rúd a változásokra, megállapítani a hődiffúziós állandót, és azt összevetni az anyagra ismert irodalmi értékkel.

A hődiffúziós egyenlet az időben változó skalármezőre, a hőmérsékletre, $T(\mathbf{r}, t)$-re, ró ki összefüggést:
$$ \frac{\partial T}{\partial t} = D \Delta T,$$
ahol $\Delta$ a Laplace-operátor, esetünkben $\frac{\partial^2}{\partial x^2}.$

A vizsgált rendszer térben és időben diszkrét, nagyjából másodpercenként méri a rúd mentén nyolc pontban a hőmérsék pillanatnyi értékét.

A hőmérő adatait írja a `hut.dat` fájlba. Amikor már írja a fájlt, adja ki az alábbi parancsot, ami bekapcsolja a rúd hűtését.

Illesszen a mérési adatokra alkalmasan választott függvényt és állapítsa meg, milyen időálladó jellemzi a rendszert! 

```python
hutes=genfromtxt('hut.dat',skip_header=27)
```

```python
def func(t,A,B,C,D):
    return A+B*math.e**(C*(t-D))
```

```python
illesztesi_parameterek_h=[]
for i in range(8):
    A=hutes[-1,i+1]
    B=hutes[100,i+1]-A
    D=1620.040
    C=1/(hutes[500,0]/1000-D)*math.log((hutes[500,i+1]-A)/B)
    popti,pcovi=curve_fit(func,hutes[100:,0]/1000,hutes[100:,i+1],p0=[A,B,C,D]) # A p0-ba egyszeruen felsoroljuk a becsult ertekeket)
    perri = np.sqrt(np.diag(pcovi))      # az illesztési paraméterek hibáinak meghatározása
    illesztesi_parameterek_h.append([popti,perri])
```

```python
fig_h, axs_h = plt.subplots(4, 2, figsize=(15,20))
fig_h.suptitle('Hűtés ábrázolása', fontsize=24)

i=1
for ax in axs_h.flat:
    ax.set(xlabel="$t$ $[s]$", ylabel="$T$ $[°C]$")
    ax.set_title('Hőmérő '+str(i),fontsize=16)
    ax.plot(hutes[100:,0]/1000,hutes[100:,i],linestyle='',marker='.', color='red', label="Hőmérő "+str(i)+" mért")
    ax.plot(hutes[100:,0]/1000,func(hutes[100:,0]/1000,*illesztesi_parameterek_h[i-1][0]),label='Hőmérő '+str(i)+' illesztett',color='black',linewidth=2)
    ax.legend(loc="upper right")
    i+=1
#A Hőmérséklet egyensúlyának beállásának ideje hőmérőnként változik: a legközelebbi (1-es) hőmérő már kb 750 s után beáll,
#az utolsó (8-as) hőmérő pedig még 1750 s után sincs teljesen egyensúlyban (később ér el oda a hatás).
```

```python
#időállandó (karakterisztikus idő)
tau_h=[]
for i in range(8):
    tau_h.append(abs(1/illesztesi_parameterek_h[i][0][2]))
print(tau_h)
#mint vártuk, a különböző hőmérők időállandója eltér, az 1-es a legkisebb és a 8-as a legnagyobb (utóbbival érdemes tovább
#számolni.)
```

Csúcsok terjedési sebessége:
Meg kell keresni az egyes csúcsok helyét, a szomszéd csúcsokat kivonni egymásból, a kapott értékek átlaga adja
a csúcsok terjedési sebességét.
Azonban nem tudtam algoritmikusan megkeresni a csúcsokat, így nincs konkrét eredményem (ránézésre kb 450 s).
A lenti függvénnyel próbálkoztam. Még olyan ötletem van, hogy illesztek függvényeket az adatpontokra, és az illesztett függvény lokális maximumait keresem meg, de erre az ötletre már nem maradt időm, ha hétfőn le akarom adni.

```python
def csucshelyek(t,T,maxszelesseg,kozepszelesseg):
    tmax=[]
    for i in range(maxszelesseg,len(t)-maxszelesseg):
        if ((T[i-1]<T[i] and T[i]>T[i+1]) and (T[i-maxszelesseg]<T[i] and T[i]>T[i+maxszelesseg]) and (T[i-kozepszelesseg]<T[i] and T[i]>T[i+kozepszelesseg])):
            tmax.append(t[i])
    return np.array(tmax)/1000

def atlagoscsucstavolsag(hullam): 
    eredmeny=[]
    for i in range(1,9):
        csucsoki=csucshelyek(hullam[:,0],hullam[:,i],4,2)
        for j in range(1,len(csucsoki)):
            eredmeny.append(csucsoki[j]-csucsoki[j-1])
    return statistics.mean(eredmeny)
```

```python
atlagoscsucstavolsag(hullam) #ez lenne maga az eredmény, de valamiért a csucshelyek függvény sokkal több (néhol meg kevesebb)
#csúcsot talált, mint kellett volna
```

# 4. Modellezés - 8 pont

Szimulálja a rendszert térben (is) finomabb rácson. Mennyire illeszkedik a szimulált adatsor a mérésekhez?

```python
def dTkoz(i,T_ti,dt): #közbülső elem hőmérséklet változása
    return -(alphasz*dA_szomszed*(2*T_ti[i]-(T_ti[i-1]+T_ti[i+1]))+alphal*dA_levego*(T_ti[i]-T0))*dt/dc
def sori(T_ti,dt):
    sork=np.array([T_ti[0]+(dtQ_fel-alphasz*dA_szomszed*(T_ti[0]-T_ti[1])-alphal*dA_levego*(T_ti[0]-T0))*dt/dc]) #0. elem hőmérséklet változása
    sorv=T_ti[-1]+(dtQ_le-alphasz*dA_szomszed*(T_ti[-1]-T_ti[-2])-alphal*dA_levego*(T_ti[-1]-T0))*dt/dc #-1. elem hőmérséklet változása
    for i in range(1,len(T_ti)-1):
        sork=np.append(sork,T_ti[i]+dTkoz(i,T_ti,dt))
    sork=np.append(sork,sorv)
    return sork
def adatsor(pontok,t,dt=0.1):
    adatsor=pontok
    for ti in range(1,len(t)):
        adatsor=np.append(adatsor,[sori(adatsor[ti-1],dt)],axis=0) #új időpillanatban vett hőmérséklet adatok
    return adatsor
```

```python
alphasz=10000 #szomszédok között a hőátvitel
alphal=500 #levegővel hőátvitel
T=T0=24
l=0.4 #rúd hossza
r=0.01 #rúd sugara
n=100 #rúd felosztása
dx=l/n #1 osztás hossza
dA_levego=2*math.pi*r*dx #levegővel vett velszín 1 darabnak
dA_szomszed=r**2*math.pi #szomszéddal vett velszín 1 darabnak
pontok=np.array([n*[T]]) #kezdő pontok hőmérséklete
t=np.linspace(0,10000,10001) #az időpontok megadása, értékük a függvényben megadott felosztástól függ
dtQ_fel=40 #fűtés Q
dtQ_le=-40 #hűtés Q
rho=9000 #rúd anyagának sűrűsége
dc=400*dA_szomszed*dx*9000 #rúd darabjának hőkapacitása
```

```python
vege=adatsor(pontok,t)
```

```python
def dTkoz(i,T_ti,dt): #közbülső elem hőmérséklet változása
    return -(alphasz*dA_szomszed*(2*T_ti[i]-(T_ti[i-1]+T_ti[i+1]))+alphal*dA_levego*(T_ti[i]-T0))*dt/dc
def sori(T_ti,dt):
    sork=np.array([T_ti[0]+(-alphasz*dA_szomszed*(T_ti[0]-T_ti[1])-alphal*dA_levego*(T_ti[0]-T0))*dt/dc]) #0. elem hőmérséklet változása
    sorv=T_ti[-1]+(-alphasz*dA_szomszed*(T_ti[-1]-T_ti[-2])-alphal*dA_levego*(T_ti[-1]-T0))*dt/dc #-1. elem hőmérséklet változása
    for i in range(1,len(T_ti)-1):
        sork=np.append(sork,T_ti[i]+dTkoz(i,T_ti,dt))
    sork=np.append(sork,sorv)
    return sork
def adatsor(pontok,t,dt=0.1):
    adatsor=pontok
    for ti in range(1,len(t)):
        adatsor=np.append(adatsor,[sori(adatsor[ti-1],dt)],axis=0) #új időpillanatban vett hőmérséklet adatok
    return adatsor
```
