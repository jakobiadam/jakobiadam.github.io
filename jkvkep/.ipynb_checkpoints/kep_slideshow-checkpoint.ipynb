# Szürke kép

A standard RGB kódolásnál, ha az intenzitást lineárisan akarjuk megkapni: $$Y_{lin}=0.2126 \cdot R_{lin} + 0.7152 \cdot G_{lin} + 0.0722 \cdot B_{lin}$$

```python
kepfalszines = imread("1kep5.png") #eredeti kép beolvasása
#szürke kép elkészítése
kepfalszurke = kepfalszines[:, :, 0]*0.2126 + kepfalszines[:, :, 1]*0.7152 + kepfalszines[:, :, 2]*0.0722
```

<img src=code1.png>

## Eredeti és szürke kép

```python
f, axs = plt.subplots(1,2,figsize=(20,20))
subplots_adjust(hspace=1)
subplot(1, 2, 1)
imshow(kepfalszines)
title("eredeti",fontsize=18)
subplot(1, 2, 2)
imshow(kepfalszurke, cmap="gray")
title("szürke",fontsize=18)
```

# Hisztogram

A feladat elkészíteni a szürke kép intenzitásának hisztogramját

```python
n, bins, patches= hist(kepfalszurke.flatten(), 256)
title("Intenzitás hisztogram", size=15)
xlabel("intenzitás")
ylabel("előfordulás")
```

# Élkeresés

Adjon meg rendre olyan kernelmátrixokat, amelyek segítségével a képen található vizszintes élek, függőleges élek vagy átlós élek emelődnek ki.

```python
kx = array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]])
ky = array([[1, 0, -1], [1, 0, -1], [1, 0, -1]])
kxy = array([[2, 1, 0], [1, 0, -1], [0, -1, -2]])
```

<img src="code3.png">

```python
imshow(ndimage.convolve(kepfalszurke, kx), cmap="gray")
title("A vízszintes élek kiemelve", size=15)
```

```python
imshow(ndimage.convolve(kepfalszurke, ky), cmap="gray")
title("A függőleges élek kiemelve", size=15)
```

```python
imshow(ndimage.convolve(kepfalszurke, kxy), cmap="gray")
title("A ferde élek kiemelve", size=15)
```
